```python
# A code for finding a reasonable value for the width of a gaussian using cross-validation.

from sklearn.cross_validation import KFold
from sklearn.neighbors.kde import KernelDensity

% matplotlib inline

def get_bins(x_range,y_range,N):

    bins_x = np.linspace(x_range[0],x_range[1],N+1)
    bins_y = np.linspace(y_range[0],y_range[1],N+1)
    
    bin_centres_x = np.array([bins_x[B] + (bins_x[B+1]-bins_x[B])/2 for B in range(N)])
    bin_centres_y = np.array([bins_y[B] + (bins_y[B+1]-bins_y[B])/2 for B in range(N)])
    
    xx,yy =  np.meshgrid(bins_x,bins_y)
    xy = np.array([np.ravel(xx),np.ravel(yy)]).T 
    
    xx_centres,yy_centres =  np.meshgrid(bin_centres_x,bin_centres_y)
    xy_centres = np.array([np.ravel(xx_centres),np.ravel(yy_centres)]).T # Used to 'score' the samples.
    
    return xx,yy,xx_centres,yy_centres,xy_centres

def f(x,y):
    return (1/math.sqrt(2*math.pi))*np.exp(-(np.sqrt(x**2+y**2))/2)
```

```python
X = np.random.randn(10**4,2) # Input dataset. We will use a simple normal distribution to test.
N=20

xx,yy,xx_c,yy_c,xy = get_bins(x_range=[-3,3],y_range=[-3,3],N=N)

kde = KernelDensity(kernel='gaussian', bandwidth=1).fit(X)
log_dens = kde.score_samples(xy).reshape(N,N)

plt.contourf(xx_c,yy_c,np.exp(log_dens))
```

```python
n_folds = 5 # Number of k-folds for the k-fold cross validation.
num_sigmas=20 # Number of sigma values to test.

x_range = [-3,3]
y_range = [-3,3]

kf = KFold(n=len(X), n_folds=n_folds,shuffle=True) # Divides the data which has n values in to n_folds sub-samples.

full_save_table = np.zeros((num_sigmas,2))

for train,test in kf:
    
    X_train,X_test = X[train],X[test]
    
    r = 0

    for sigma in np.linspace(0.01,2,num_sigmas):
    
        kde = KernelDensity(kernel='gaussian', bandwidth=sigma).fit(X_train)
        
        log_dens = kde.score_samples(xy).reshape(N,N)
        dens = np.exp(log_dens)
        
        train_dens = np.histogram2d(X_test[:,0],X_test[:,1],bins=N,range=[x_range,y_range]
            ,normed=True)[0]
        
        diff = np.sum((train_dens-dens)**2)
        
        full_save_table[r,1] = full_save_table[r,1] + [diff]
        r = r+1

full_save_table[:,0] = np.linspace(0.01,2,num_sigmas)

plt.plot(full_save_table[:,0],full_save_table[:,1],'k--')
plt.plot(full_save_table[np.argmin(full_save_table[:,1]),0],
         full_save_table[np.argmin(full_save_table[:,1]),1],'ro')

plt.xlabel('gaussian bandwidth')
plt.ylabel('Total residual')
```

```python
min_value = full_save_table[np.argmin(full_save_table[:,1]),0]

H,xe,ye = np.histogram2d(X[:,0],X[:,1],range=[x_range,y_range],bins=20,normed=True)

xg,yg = np.meshgrid(xe,ye)

plt.contour(H,extent=(xe[0],xe[-1],ye[0],ye[-1]),
            levels=(0.03,0.06,0.12),colors='k',linewidths=2,label="'Optimised' KDE")

kde = KernelDensity(kernel='gaussian', bandwidth=min_value).fit(X)
log_dens = kde.score_samples(xy).reshape(N,N)
dens = np.exp(log_dens)

plt.contour(xx_c,yy_c,np.exp(log_dens),colors='r',linewidths=2,linestyles='dashed',
            levels=(0.03,0.06,0.12),label='Fitted distribution')

plt.legend()
```
