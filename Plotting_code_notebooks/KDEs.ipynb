```python
# A code for finding a reasonable value for the width of a gaussian using cross-validation.

from sklearn.cross_validation import KFold
from sklearn.neighbors.kde import KernelDensity

% matplotlib inline
```

```python
X = np.random.randn(10**3)[:,np.newaxis] # Input dataset. We will use a simple normal distribution to test.

def get_bins(rng,N):

    bins = np.linspace(rng[0],rng[1],N+1)
    bin_centres = np.array([bins[B] + (bins[B+1]-bins[B])/2 for B in range(N)])
    
    return bins,bin_centres
```

```python
def f(x):
    return (1/math.sqrt(2*math.pi))*np.exp(-(x**2/2))
```

```python
N = 25 # Number of bins to histogram the sample by.
rng = [-3,3] # Range of the histogram.
n_folds = 5 # Number of k-folds for the k-fold cross validation.
num_sigmas=50 # Number of sigma values to test.

bins,bin_centres = get_bins(rng,N)
X_plot = bin_centres[:, np.newaxis] 

kf = KFold(n=len(X), n_folds=n_folds,shuffle=True) # Divides the data which has n values in to n_folds sub-samples.

full_save_table = np.zeros((num_sigmas,2))

for train,test in kf:
    
    X_train,X_test = X[train],X[test]
    
    r = 0

    for sigma in np.linspace(0.01,2,num_sigmas):
    
        kde = KernelDensity(kernel='gaussian', bandwidth=sigma).fit(X_train)
        
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        train_dens = np.histogram(X_test,bins=bins,normed=True)[0]
        
        diff = np.sum((train_dens-dens)**2)
        
        full_save_table[r,1] = full_save_table[r,1] + [diff]
        r = r+1

full_save_table[:,0] = np.linspace(0.01,2,num_sigmas)

plt.plot(full_save_table[:,0],full_save_table[:,1],'k--')
plt.plot(full_save_table[np.argmin(full_save_table[:,1]),0],
         full_save_table[np.argmin(full_save_table[:,1]),1],'ro')

plt.xlabel('gaussian bandwidth')
plt.ylabel('Total residual')
#plt.show()
```

```python
min_value = full_save_table[np.argmin(full_save_table[:,1]),0]

X_plot2 = np.linspace(rng[0],rng[1],1000)[:, np.newaxis] 

plt.hist(X,bins=bins,histtype='stepfilled',color='y',normed=True)

kde = KernelDensity(kernel='gaussian', bandwidth=min_value).fit(X)
log_dens = kde.score_samples(X_plot2)

plt.plot(X_plot2,np.exp(log_dens),'g-',linewidth=3,label="'Optimised' KDE")
plt.plot(X_plot2,f(X_plot2),'k--',linewidth=2,label="'True' probability")

plt.xlabel('$x$')
plt.ylabel('$f(x)$')

plt.legend()
```
