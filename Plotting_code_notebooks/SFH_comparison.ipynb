```python
colours=["orange","red","purple","green","blue"]
cmaps=["YlOrBr","Reds","BuPu","Greens","Blues"]

levels = np.array([5,10,15])*(10**(-4))

def load_assign(cx,cy):

    table,full_table = load_data.load(cx=cx,cy=cy,p_th=0.5,N_th=10,norm=False,p_values="d")
    bins,table = load_data.assign(table=table,Nb=20,th=0.5,equal_samples=True,
        redistribute=False,rd_th=0,ct_th=0,print_sizes=False)
    
    return bins,table

def plot_taus(cx,cy,colour1,colour2,xlim,ylim,dust_val):
    
    bins,table = load_assign(cx,cy)

    for m in [1,4]:
    
        tr = table[bins[:,1] == m]
    
        plotting.contour(table=tr,xlims=xlim,ylims=ylim,grid_spacing=50
            ,contour_type="contourf+line",ax=ax,alpha=0.5,colour=[colours[m],cmaps[m]]
            ,levels=levels,sigma=False)
    
    sm_plot.plot_line(vary="tau",taus=[0.1,0.2,1],Avs=[0],dust_arrow=dust_val,
        colour1=colour1,colour2=colour2)

    plt.xlabel("${{{}}}".format(colour1[0]) + " - {{{}}}$".format(colour1[1]))
    plt.ylabel("${{{}}}".format(colour2[0]) + " - {{{}}}$".format(colour2[1]))
    
    return None

def plot_dusts(cx,cy,colour1,colour2,xlim,ylim,tau_val):
    
    bins,table = load_assign(cx,cy)

    for m in [1,4]:
    
        tr = table[bins[:,1] == m]
    
        plotting.contour(table=tr,xlims=xlim,ylims=ylim,grid_spacing=50
            ,contour_type="contourf+line",ax=ax,alpha=0.5,colour=[colours[m],cmaps[m]]
            ,levels=levels,sigma=False)
    
    sm_plot.plot_line(vary="Av",taus=[tau_val],Avs=[0,0.4,0.8],dust_arrow=None,
        colour1=colour1,colour2=colour2)

    plt.xlabel("${{{}}}".format(colour1[0]) + " - {{{}}}$".format(colour1[1]))
    plt.ylabel("${{{}}}".format(colour2[0]) + " - {{{}}}$".format(colour2[1]))
    
    return None
```
