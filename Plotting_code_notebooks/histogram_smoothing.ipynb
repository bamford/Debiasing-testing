```python
import plotting
from prefig import Prefig
from scipy.optimize import curve_fit
import scipy.ndimage
from matplotlib.colors import LogNorm

Prefig()

% matplotlib inline

relative_size = 2
########################################
figure_size_2x3 = [8,5]
figure_size_2x3 = [x*relative_size for x in figure_size_2x3]

font_size_2x3 = (10)*relative_size

########################################
figure_size_6x1 = [5,8]
figure_size_6x1 = [x*relative_size for x in figure_size_6x1]

font_size_6x1 = (10)*relative_size
```

```python
def con(D,xlims,ylims,N):
    
    xedges=np.linspace(xlims[0],xlims[1],N+1)
    yedges=np.linspace(ylims[0],ylims[1],N+1)

    extent=[xedges[0],xedges[-1],yedges[0],yedges[-1]]

    H,xi,yi=np.histogram2d(D[:,-2],D[:,-1], bins=(yedges, xedges),normed=True)
    #H=scipy.ndimage.interpolation.zoom(H,2,order=1)
    H=scipy.ndimage.filters.gaussian_filter(input=H,sigma=0.1,order=0)
    
    return H,extent,xedges,yedges

#----------------------------------------------------------------------------------------------------------------------------------
def line_f(x,a,b):
    return a*x+b

#----------------------------------------------------------------------------------------------------------------------------------
def contour(cx,cy,grid,f,ps,con_full,plot_line,xlims,ylims,f_size,levels):
  
    table,full_table = plotting.load_data(cx=cx,cy=cy,p_th=0.5,N_th=10,norm=False,p_values="d")
    bins,table = plotting.assign(table=table,Nb=20,th=0.5,bin_type="equal samples",redistribute=False,rd_th=0,ct_th=0)
    
    C = ["purple","red","magenta","green","blue","orange"]
    
    x_ex = np.array(xlims)
    
    N_c = 100
    
    if con_full == 1:
        H,extent,xedges,yedges=con(full_table,N=N_c,xlims=xlims,ylims=ylims)
        po,pc=curve_fit(line_f,full_table[:,-1],full_table[:,-2])

        for a in range(6):
            ax=ps[a]
            rng=np.array([xlims,ylims])
            #ax.contourf(H,extent=extent,alpha=0.5,cmap="Greys",levels=np.linspace(0,1000,100))
            ax.hist2d(full_table[:,-1], full_table[:,-2], bins=N_c,cmap="Greys",range=rng,norm=LogNorm())
            ax.imshow(H,extent=extent,cmap="Greys")
            
            #H,extent,xedges,yedges=con(full_table,N=100,xlims=xlims,ylims=ylims)
            po,pc=curve_fit(line_f,full_table[:,-1],full_table[:,-2])

        for a in range(6):
            ax=ps[a]
            rng=np.array([xlims,ylims])

            #X, Y = np.meshgrid(xedges, yedges)
            #ax.pcolor(X, Y, np.log(H),cmap="Greys")
            #ax.set_aspect('auto')
                
            if plot_line == 1:
                ax.plot(x_ex,line_f(x_ex,po[0],po[1]),linewidth=2,color="black")
    
    for a in range(6):
            
        ax=ps[a]
        t_select=table[bins[:,1] == a]

        H,extent,xedges,yedges=con(t_select,N=grid,xlims=xlims,ylims=ylims)
        #c = ax.contour(H,extent=extent,colors=C[a],linewidths=2,norm=True,levels=levels)
        #ax.clabel(c, inline=1, fontsize=10)
        
        if plot_line == 1:
            po,pc=curve_fit(line_f,t_select[:,-1],t_select[:,-2])
            ax.plot(x_ex,line_f(x_ex,po[0],po[1]),linewidth=2,color=C[a])
            
    # Add a colourbar        
    extent_c=[0.9, 0.08, 0.02, 0.9]
    
    f.subplots_adjust(right=extent_c[0]-0.01)
    plt.hist2d(full_table[:,-1], full_table[:,-2],bins=N_c,cmap="Greys",range=rng,norm=LogNorm())

    cbar_ax = f.add_axes(extent_c)  
    plt.colorbar(cax=cbar_ax) 
    
    f.text(0.98,(extent_c[1]+extent_c[3])-(extent_c[3]/2), "$N_{gal}$", 
           ha='center', va='center', rotation='vertical',size=f_size)

    for a in range(6):
        ax.set_xlim(xlims)
        ax.set_ylim(ylims)
    
    return None
```

```python
# Colour magnitude diagram 1

fig,ps=plotting.make_2x3_plot(xlims=[-23,-21],xticks=[-22.5,-22,-21.5,-21],ylims=[1,3],yticks=[1.5,2,2.5,3],
    x_label="$M_r$",y_label="u-r",title_position=[0.02,0.95],f_size=font_size_2x3,fig_size=figure_size_2x3)

#plotting.histogram(cx=["PETROMAG_MR"],Nb=20,bin_extent=[-23,-21],ps=ps,full_hist=True)
contour(cx=["PETROMAG_MR"],cy=["PETROMAG_MU","PETROMAG_MR"],grid=30,
    f=fig,ps=ps,con_full=True,plot_line=True,xlims=[-23,-21],ylims=[1,3],f_size=font_size_2x3,levels=np.linspace(0.3,1.2,5))

#plt.savefig("figs/cm.jpg")
```
